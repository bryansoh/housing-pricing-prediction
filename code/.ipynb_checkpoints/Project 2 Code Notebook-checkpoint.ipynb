# Background
A real estate investment company ideally buys houses at low prices and sell them high for a profit. As a data scientist working at such a company based in Ames, Iowa, we are tasked to build a machine learning model to predict the true value of a house. Using the model predictions as a reference, the company will then be able to decide if a house is underpriced or overpriced.

The data used is the [Ames housing dataset][1] that was compiled by Dean De Cock. It describes the sale of individual residential property in Ames, Iowa from 2006 to 2010.

[1]: http://jse.amstat.org/v19n3/decock.pdf "Title"

# Problem Statement
Using a regression model, predict the true value of a house in Ames, Iowa with an RMSE of less than $20K.

Identify features of a house with the highest positive impact on its SalePrice.

Identify features of a house with the highest negative impact on its SalePrice.

# 3) Data Modeling

```python
X = train.drop(columns = 'Natural Log of SalePrice')
y = train['Natural Log of SalePrice']
```

```python
X.head()
```

### Train/test split at 75% (train) to 25% (test) to check model performance

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)
ss = StandardScaler()
X_train_ss = ss.fit_transform(X_train)
X_test_ss = ss.transform(X_test)
X_ss = ss.fit_transform(X)
```

```python
lr = LinearRegression()
lr.fit(X_train_ss, y_train)
print(f" Linear Regression's train score is {lr.score(X_train_ss, y_train)}")
print(f" Linear Regression's test score is {lr.score(X_test_ss, y_test)}")
```

### Although Linear Regression's train score is high at 95%, its test score is negative and hence the model is performing badly when generalizing to new data. 

### This is expected as Linear Regression is a simple regression model. While it is not likely to perform better than other regression models, it still functions as a good starting point.

### Next, a Ridge Regression model will be evaluated. This type of model contains a penalty term alpha that shrinks the coefficient of features that do not contribute much to the dependent variable being predicted.

```python
ridge_alphas = np.logspace(0, 5, 200)
ridge = RidgeCV(alphas=ridge_alphas, cv=5)
ridge.fit(X_train_ss, y_train)
ridge_train_score = ridge.score(X_train_ss, y_train)
ridge_test_score = ridge.score(X_test_ss, y_test)
ridge_train_rmse = mean_squared_error(np.exp(y_train), np.exp(ridge.predict(X_train_ss)), squared=False)
ridge_test_rmse = mean_squared_error(np.exp(y_test), np.exp(ridge.predict(X_test_ss)), squared=False)
print(f" Ridge Regression's train score is {ridge_train_score}")
print(f" Ridge Regression's test score is {ridge_test_score}")
print(f" Ridge Regression's train RMSE is ${ridge_train_rmse}")
print(f" Ridge Regression's test RMSE is ${ridge_test_rmse}")
```

### As expected, the Ridge Regression model performs better than the Linear Regression model. Next, a Lasso Regression model is evaluated. This type of model also contains a penalty term alpha, except coefficients of features that do not contribute much can actually be zero'd out.

```python
lasso = LassoCV(n_alphas=500, cv=5)
lasso.fit(X_train_ss, y_train)
lasso_train_score = lasso.score(X_train_ss, y_train)
lasso_test_score = lasso.score(X_test_ss, y_test)
lasso_train_rmse = mean_squared_error(np.exp(y_train), np.exp(lasso.predict(X_train_ss)), squared=False)
lasso_test_rmse = mean_squared_error(np.exp(y_test), np.exp(lasso.predict(X_test_ss)), squared=False)
print(f" Lasso Regression's train score is {lasso_train_score}")
print(f" Lasso Regression's test score is {lasso_test_score}")
print(f" Lasso Regression's train RMSE is ${lasso_train_rmse}")
print(f" Lasso Regression's test RMSE is ${lasso_test_rmse}")
```

### Lastly, an ElasticNet Regression model is evaluated. This type of model incorporates elements from both Ridge and Lasso. In addition to the penalty term alpha, it contains a term called the l1 ratio that ranges from 0 to 1. If l1 ratio equals to 1, the ElasticNet Regression model becomes a Lasso Regression model. If l1 ratio equals to 0, it becomes a Ridge Regression model. 

```python
l1_ratios = np.linspace(0.01, 1.0, 25)
enet = ElasticNetCV(l1_ratio=l1_ratios, n_alphas=100, cv=5)
enet.fit(X_train_ss, y_train)
enet_train_score = enet.score(X_train_ss, y_train)
enet_test_score = enet.score(X_test_ss, y_test)
enet_train_rmse = mean_squared_error(np.exp(y_train), np.exp(enet.predict(X_train_ss)), squared=False)
enet_test_rmse = mean_squared_error(np.exp(y_test), np.exp(enet.predict(X_test_ss)), squared=False)
print(f" ElasticNet Regression's train score is {enet_train_score}")
print(f" ElasticNet Regression's test score is {enet_test_score}")
print(f" ElasticNet Regression's train RMSE is ${enet_train_rmse}")
print(f" ElasticNet Regression's test RMSE is ${enet_test_rmse}")
```

```python
model_performance = [['Ridge', ridge_train_score, ridge_test_score, ridge_train_rmse, ridge_test_rmse], ['Lasso',lasso_train_score, lasso_test_score, lasso_train_rmse, lasso_test_rmse], ['ElasticNet',enet_train_score, enet_test_score, enet_train_rmse, enet_test_rmse]]
model_performance_df = pd.DataFrame(model_performance, columns = ['Model Type', 'Train Score', 'Test Score', 'Train RMSE', 'Test RMSE'] )
model_performance_df
```

### The objective of the model is to predict the true value of a given house. Hence, how well the model generalizes to new data is very important. With that, the 2 results to pay attention to are Test Score (the higher the better) and Test RMSE (the lower the better). 

### Observations:
- Although the Ridge model has the highest Train Score, it also has the lowest Test Score. In other words, it is the worst at generalizing to new data.
- Although the Lasso model has higher Train (0.0206% higher) and Test (0.0127% higher) Scores than the ElasticNet model, the ElasticNet model has a lower Test RMSE (0.425% lower).

### Insights:
- Both Lasso and ElasticNet have very similar Test Scores. Generally, they perform to the same level when generalizing to new data. 
- Between Lasso and ElasticNet, the difference in Test RMSE is more significant than the difference in Test Scores. 0.425% vs 0.0127% difference.
- The objective of this model is to drive profit for the company by identifying the true value of a house and hence purchase it at a low price. Since the Test Score is so similar, we should look at the model with the lower RMSE to ensure our predictions in general are closer to the true value.
- Hence, ElasticNet is the best performing model.

## 4) Conclusion and Insights:

### Coefficient Interpretation:
- As the model is predicting the natural log of SalePrice, every unit increase in the feature will result in an increase of ((exp_coef  1) * 100) percent of SalePrice, as reflected in the bar graph above.
- The above is true for all features except Lot Area because the natural log of Lot Area was used in the model.
- The coefficient of Natural Log of Lot Area will be interpreted as the percent increase of SalePrice per 1% increase in Lot Area.

### These are the features with the highest positive impact on SalePrice:
- Gr Liv Area: SalePrice increase of 11.39% per square foot increase of Above grade (ground) living area square feet
- Overall Qual: SalePrice increase of 7.36% per improvement in rating of the overall material and finish of the house
- Year Built: SalePrice increase of 5.71% per year later a house is built
- Lot Area: SalePrice increase of 4.4% per 1% increase in lot area 
- Overall Cond: SalePrice increase of 3.93% per improvement in rating of the overall condition of the house

### Although these features have the highest SalePrice increase per unit increase, thinking logically, continuous features will have the most impact on SalePrice compared to ordinal and categorical features as they have a larger capacity to increase.

### Hence, the 5 most impactful features are:
- Gr Liv Area: Increase of 11.39% per square foot increase of Above grade (ground) living area square feet
- Lot Area: Increase of 4.4% per 1% increase in lot area 
- BsmtFin Type 1: Increase of 2.8% per square foot increase in Type 1 finished square feet
- Total Bsmt SF: Increase of 2.74% per square foot increase in total basement area
- Garage Area: Increase of 1.04% per square foot increase in garage size

### These are the 5 features with the highest negative impact on SalePrice:
- Heating_Grav: SalePrice decrease of 1.406% when the house has a gravity furnace for heating
- MS Zoning_C (all): SalePrice decrease of 1.405% when the house is in a commercial zone
- KitchenAbvGr: SalePrice decrease of 1.058% per increase in number of Kitchens above grade
- MS Zoning_RM: SalePrice decrease of 0.938% when the house is in a residential zone with medium density
